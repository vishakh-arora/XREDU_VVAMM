Dataset Description: This dataset contains the number of calories burned by a person while performing some activity/exercise.
It currently contains 248 activities and exercises ranging from running, cycling, calisthenics, etc.  
Data Source: https://www.kaggle.com/aadhavvignesh/calories-burned-during-exercise-and-activities

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("exercise_dataset.csv")
df
```

```python
#exercises that burn the most calories
highest_cal = df.sort_values('Calories per kg', ascending=False).head()
highest_cal
```

```python
plt.figure(figsize=(15,5))
plt.bar(highest_cal["Activity, Exercise or Sport (1 hour)"], highest_cal["Calories per kg"])
plt.xlabel("Exercise")
plt.ylabel("Calories burned per kg")
plt.title("Bar graph for calories per kg for the highest burning exercises")

plt.show()
```

```python
#exercises that burn the least calories
lowest_cal = df.sort_values('Calories per kg', ascending=True).head()
lowest_cal
```

```python
plt.figure(figsize=(15,5))
plt.bar(lowest_cal["Activity, Exercise or Sport (1 hour)"], highest_cal["Calories per kg"])
plt.xlabel("Exercise")
plt.ylabel("Calories burned per kg")
plt.title("Bar graph for calories per kg for the lowest burning exercises")

plt.show()
```

```python
combined_df = lowest_cal.head(3).append(highest_cal.head(3))
combined_df
```

```python
plt.figure(figsize=(20,5))
plt.bar(combined_df["Activity, Exercise or Sport (1 hour)"], combined_df["Calories per kg"])
plt.xlabel("Exercise")
plt.ylabel("Calories burned per kg")
plt.title("Bar graph for calories per kg for the lowest and highest burning exercises")

plt.show()
```
